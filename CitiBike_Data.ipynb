### Clean Up CitiBike Data

```python
# Import csv files
jan_2019 = "JC-201901-citibike-tripdata.csv"
feb_2019 = "JC-201902-citibike-tripdata.csv"
mar_2019 = "JC-201903-citibike-tripdata.csv"
apr_2019 = "JC-201904-citibike-tripdata.csv"
may_2019 = "JC-201905-citibike-tripdata.csv"
jun_2019 = "JC-201906-citibike-tripdata.csv"

jan_2020 = "JC-202001-citibike-tripdata.csv"
feb_2020 = "JC-202002-citibike-tripdata.csv"
mar_2020 = "JC-202003-citibike-tripdata.csv"
apr_2020 = "JC-202004-citibike-tripdata.csv"
may_2020 = "JC-202005-citibike-tripdata.csv"
jun_2020 = "JC-202006-citibike-tripdata.csv"
```

```python
# Create dataframes for each csv file
jan2019_df = pd.read_csv(jan_2019)
feb2019_df = pd.read_csv(feb_2019)
mar2019_df = pd.read_csv(mar_2019)
apr2019_df = pd.read_csv(apr_2019)
may2019_df = pd.read_csv(may_2019)
jun2019_df = pd.read_csv(jun_2019)

jan2020_df = pd.read_csv(jan_2020)
feb2020_df = pd.read_csv(feb_2020)
mar2020_df = pd.read_csv(mar_2020)
apr2020_df = pd.read_csv(apr_2020)
may2020_df = pd.read_csv(may_2020)
jun2020_df = pd.read_csv(jun_2020)
```

```python
# Merge all dataframes into one to combine for analysis

data1 = pd.merge(jan2019_df, feb2019_df, how="outer")
data2 = pd.merge(data1, mar2019_df, how="outer")
data3 = pd.merge(data2, apr2019_df, how="outer")
data4 = pd.merge(data3, may2019_df, how="outer")
data5 = pd.merge(data4, jun2019_df,how="outer")
data6 = pd.merge(data5, jan2020_df, how="outer")
data7 = pd.merge(data6, feb2020_df, how="outer")
data8 = pd.merge(data7, mar2020_df, how="outer")
data9 = pd.merge(data8, apr2020_df, how="outer")
data10 = pd.merge(data9, may2020_df, how="outer")
data = pd.merge(data10, jun2020_df, how="outer")
```
